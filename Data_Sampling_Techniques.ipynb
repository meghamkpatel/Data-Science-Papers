## Why is data sampling important?
Data sampling is important for several reasons in the fields of statistics and data analysis:

* **Resource Conservation:** Sampling can be essential in situations where data collection is resource-intensive or intrusive. For instance, in medical research, it may be impractical to collect a complete, comprehensive history of patients, so a sample of cases is used for analysis.

* **Exploratory Data Analysis:** In exploratory data analysis, data sampling is performed as the first step to provide an initial understanding of the dataset's characteristics and thus helping analysts decide on the most appropriate statistical methods and visualizations.

* **Statistical Inference:** Sampling allows statisticians to make inferences about a population based on the characteristics of a sample. Well-sampled and representative data can lead to accurate population-level estimates.

* **Machine Learning:** In machine learning, models can be trained on smaller samples to speed up the model development process while also producing similar results. Researchers can iterate runs more quickly to fine-tune algorithms and model parameters.

* **Testing and Validation:** In A/B testing or hypothesis testing, data sampling is used to test accuracy of the model. It allows researchers to draw conclusions about the population based on the performance or behavior of the sample.

* **Privacy and Security:** Organizations can reduce the risk of exposing sensitive or confidential data while still gaining insights by working with a sample.

##Advantages of data sampling
* **Handling Large Datasets:** Many real-world datasets are massive and may contain millions or even billions of data points. To analyze the entire dataset would be computationally intensive and time-consuming. Data sampling allows analysts to work with a smaller, manageable subset of the data without sacrificing the validity of the analysis.

* **Cost-Efficiency:** Collecting, storing, and processing large datasets can be expensive, both in terms of hardware and computational resources. By sampling the data, organizations can save on storage and processing costs while still gaining valuable insights.

* **Faster Analysis:** In scenarios where timely decisions are required, working with a smaller sample enables quicker data analysis and experimentation. This is particularly useful for real-time analytics or decision support systems.

* **Accuracy:** Researchers can accurately interpret information about the total population by selecting a representative sample if the correct techniques are used.

* **Bias elimination:** Sampling can help to eliminate bias in data analysis,when well-designed, by limiting the influence of outliers, errors and other kinds of bias that may undermine the analysis of the entire population.

* **Reducing Noise:** Some datasets may contain noisy or irrelevant data points that can distort analysis results. Carefully sampling can help reduce the impact of noise on the analysis.


##Challenges of data sampling
* **Risk of bias:** One of the main challenges with data sampling is the possibility of introducing bias into the sample. By chosing the wrong technique to sample the population, we can assume incorrect or misleading conclusions and misrepresent the population.
* **Determining the sample size:** Determining an appropriate sample size is the most difficult part in data sample and directly impacts the results of the analysis. If the sample size is too small, the results might not be accurate since the sample will not be representative of the population.
* **Sampling error:** Sampling error is the discrepancy between the sample and the population. Variances in the accuracy of the results may be affected by chance, bias or other factors.
* **Sampling method:** The choice of sampling method can vary depending on the research question and population being studied. However, selecting the appropriate sampling technique is vital because different techniques result in  different distributions and assume specific circumstances.

##Data sampling process
The process of data sampling typically involves the following steps:

1. **Defining the population:** The population is the entire set of data from which the sample is drawn. To guarantee that the sample is representative of the entire population, the target population must be precisely defined, including all essential traits and criteria.
2. **Selecting a sampling technique:** The next step is to choose the best sampling method based on the research question and the characteristics of the population under study. There are several methods for drawing samples from data such as simple random sampling, cluster sampling, stratified sampling and systematic sampling (discussed below)
3. **Determining the sample size:** The optimum sample size required to produce accurate and reliable results should be chosen. This decision is usually influenced by certain factors, such as money, time constraints, and precision. The sample size should be large enough to be representative of the population, but not so large that it becomes impractical to work with.
4. **Collecting the data:** The data is collected from the sample using the sampling approach that was chosen, such as interviews, surveys or observations. This may entail random selection or separation by groups, depending on the research question.
5. **Analyzing the sample data:** After collecting the data sample, it's processed and analyzed to draw conclusions about the population. The results of the analysis are then modeled or applied to the entire population.

##Data Sampling Methods
While there are many different methods for drawing samples from data, we will discuss probability sampling and how to implement it in this notebook.
Probability sampling is an approach that utilizes randomness to choose points in the data set to ensure that there is no correlation between points in the sample set. Variations in probability sampling include the following:

* **Random Sampling:** (Simple Random Sampling): In this method, every data point in the dataset has an equal chance of being selected. It's like picking names out of a hat or rolling a dice. This is the most straightforward form of sampling.

* **Stratified Sampling:** In stratified sampling, the population is divided into subgroups or strata based on certain characteristics (e.g., age groups, geographical regions). Then, random sampling is performed independently within each stratum. This ensures that each subgroup is represented in the sample and guarantee a simular distribution as the population distribution.

* **Systematic Sampling:** Systematic sampling involves selecting every nth data point from the dataset. For example, if there are 1,000 data points and a sample of 100 points are needed, every 10th data point can be selected.

* **Cluster Sampling:** In cluster sampling, the population is divided into clusters based on some natural grouping, such as geographical regions, schools, cities, or neighborhoods. Then, a random sample of clusters is selected and all data points within the selected clusters are included in the sample.

* **Bootstrapping:** Bootstrapping is a resampling technique where data points are randomly selected with replacement from the dataset. It's often used for estimating the sampling distribution of a statistic, such as the mean or standard deviation.

# Simple Random Sampling

Simple random sampling is a widely used technique in statistics and research when to perform preliminay exploration of the dataset to gain initial insights or determine the need for more complex sampling methods. It is best when the population has little variation between individuals or elements. Random sampling ensures that every unit has an equal opportunity to be selected in a subset of data or individuals from a larger population for analysis. This technique is employed primarily in situations where there is limited prior knowledge about the population, a lack of specific characteristics to stratify by, and a need to ensure an unbiased and representative sample.

image.png


We will use the ```sample()``` function from the pandas library to randomly select 100 values from the credit table. We choose to set the random_state as  None because calling the function multiple times will reuse the same instance and will produce different results.

Parameters:
* n: the number of items in the sample
* replace: whether the elements of the table can be inserted into the sample more than once
* random_state: determines random number generation for shuffling the data. (default is None).



````python
df_sample_random = df.sample(n = 100, replace = False, random_state = None)
df_sample_random.describe()
````

````python
fig = plt.figure()
plt.figure(figsize=(100,50))
df2 = pd.concat({'df': df.melt(), 'df_sample_random': df_sample_random.melt()}, names=['source', 'old_index'])
df2 = df2.reset_index(level=0).reset_index(drop=True)

t = sns.catplot(data=df2, kind='box', col='variable', y='value', x='source', palette='turbo',hue='source', legend = 'brief', height=4, aspect=0.8, col_wrap=6, sharey=False)
t.set(xlabel='', xticks=[]) # remove x labels and ticks
t.add_legend()
````

Notice how the mean and standard deviation for the sampled data are slightly higher than the complete dataset. When we plot both the dataset and the sample as a boxplot on top of each other, we see that the income and loan have different quartiles. Looking specifically at the default attribute, we see that the statistics are similar but too small to plot on the boxplot. We can expect this sample to have some inaccuracies when predicting defaults based on the income and loan attributes.

While simple random sampling has its advantages, it's important to recognize its limitations as well. It may not be the best choice when there is a pattern of subgroupings or clusters identified, or when the cost of data collection is a significant concern. In such cases, more advanced sampling methods like stratified sampling or cluster sampling may be more appropriate.

# Stratified Sampling


Stratified sampling is a valuable sampling technique used when you want to ensure that your sample accurately represents different subgroups or strata within a population. It is particularly useful in situations where the population has inherent variability or when you want to study specific subpopulations separately. By focusing on each subgroup, we can reduce the overall variability in estimates compared to simple random sampling.

image.png

To perform stratified sampling effectively, follow these steps:

1. Identify the relevant subgroups or strata within your population.
2. Determine the sample size or proportion you want to allocate to each stratum. This can be based on the stratum's size in the population or other criteria.
3. Randomly sample from each stratum independently. You can use techniques like simple random sampling or systematic sampling within each stratum.
4. Combine the samples from each stratum to form your final stratified sample.

Let's revisit the credit dataset.

````python
display(df.describe())
````

````python
# import sklearn's Stratified library
from sklearn.model_selection import StratifiedShuffleSplit as ss
````

We can create a function to determine whether the data is proportional based on a pre-defined variable, in this case, the default attribute. We can see that the amount of non-defaults in the dataset is disproportionately higher than the amount of defaults.

````python
def counts(df, var):
    c = df[var].value_counts(dropna=False)
    p = df[var].value_counts(dropna=False, normalize=True)
    return pd.concat([c,p], axis=1, keys=['counts', '%'])

display(counts(df, 'default'))
````

To extract stratified samples easily, we can create a function with parameters for size and the variable that we stratify by.

````python
# function for extracting stratified samples with useful parameters

def sample_stratified(dataframe, size, variable):
    sp = ss(test_size=size)
    for _, y in sp.split(dataframe, dataframe[variable]):
        dataframey = dataframe.iloc[y]
    return dataframey
````

Now we can split the credit dataset by a proportion allocate to each stratum and randomly sample from each stratum independently. In this example, we choose to split the dataset by 10%.

````python
# using the function

df_sample_stratified = sample_stratified(df, 0.1, 'default')

display(df_sample_stratified.describe())

display(counts(df_sample_stratified,'default'))
print('\n')
````

````python
fig = plt.figure()
plt.figure(figsize=(100,50))
df2 = pd.concat({'df': df.melt(), 'df_sample_stratified': df_sample_stratified.melt()}, names=['source', 'old_index'])
df2 = df2.reset_index(level=0).reset_index(drop=True)

t = sns.catplot(data=df2, kind='box', col='variable', y='value', x='source', palette='turbo',hue='source', legend = 'brief', height=4, aspect=0.8, col_wrap=6, sharey=False)
t.set(xlabel='', xticks=[]) # remove x labels and ticks
t.add_legend()
````

Based on the statistics and box plot for the stratified sample and the original credit data, we see that there is very little difference between the ranges of each attribute, showing us that our sample has an even distribution of all groups. The mean for the incomes varies slightly and may cause some errors when predicting defaults.

As we saw, stratified sampling is a powerful method for improving the accuracy and representativeness of your sample when dealing with diverse populations or when you want to conduct subgroup-specific analyses.

# Systematic Sampling


Systematic sampling is a method of data sampling where you select every nth item from a list or population after randomly selecting a starting point. This approach is particularly useful when dealing with a large dataset and where we can assume that the data is somewhat uniformly distributed throughout the population.

Systematic sampling may not be the best choice if there are hidden patterns such as clustering or if the distribution is not uniform.

To perform systematic sampling, follow these steps:
1. Determine the Sample Size
2. Calculate the Sampling Interval
3. Select a Random Starting Point
4. Sample Selection
5. Data Collection

image.png

Let's revisit the credit dataset.

````python
display(df.describe())
````

In a function, let's take the number of records in the dataset and divide it by the sample size we want to generate to create a selection range of the elements of our sample. The selection process must start randomly where the first element must be selected at a random position in the first range. Then we obtain sample indexes systematically based on the interval calculated. We choose to keep the random_state as None because calling the function multiple times will reuse the same instance and will produce different results.

Parameters:
* sample_size: number of elements to choose from the dataset
* state: determines random number generation for shuffling the data. (default is None).

````python
# function for extracting systematic samples based on sample size and state

def sample_systematic(dataframe, sample_size, state = None):
    length = len(df)
    interval = length // sample_size
    rd.seed(state)
    first = rd.randint(0, interval)
    indexes = np.arange(first, length, step = interval)
    return dataframe.iloc[indexes]
````

Implementing our function, we will now choose a sample size of 100 and view the statistics of the resulting sample set.

````python
# using the function

df_sample_systematic = sample_systematic(df, 100)
display(df_sample_systematic.describe())
````

````python
fig = plt.figure()
plt.figure(figsize=(100,50))
df2 = pd.concat({'df': df.melt(), 'df_sample_systematic': df_sample_systematic.melt()}, names=['source', 'old_index'])
df2 = df2.reset_index(level=0).reset_index(drop=True)

t = sns.catplot(data=df2, kind='box', col='variable', y='value', x='source', palette='turbo',hue='source', legend = 'brief', height=4, aspect=0.8, col_wrap=6, sharey=False)
t.set(xlabel='', xticks=[]) # remove x labels and ticks
t.add_legend()
````

Based on the statistics and box plot for the systematic sample and the original credit data, we see that there is very little difference between the ranges of each attribute. The results look similar to the stratified sampling results but we can assume some sampling error due to the varied mean and standard dev for loan, age, and income attributes.

In summary, systematic sampling is a valuable method when you need to efficiently select a representative sample from a large dataset, and you can assume some level of uniform distribution. However, it's important to be cautious of hidden patterns or biases that may arise in specific situations, and to carefully select a random starting point to maintain randomness.

# Cluster Sampling


Cluster sampling is a sampling technique used in situations where the population is naturally grouped into clusters or subgroups, and it may be impractical or too costly to sample from the entire population. Instead of selecting individual elements from the population, cluster sampling involves randomly selecting entire clusters and then observing the distribution within those clusters.

When the population naturally forms clusters, either geographically or by some other criterion, it makes sense to use cluster sampling. Cluster sampling is particularly effective in cases like clinical trials or experiments where the intervention or treatment is administered at the cluster level, such as schools, communities, or hospitals. In some cases, cluster sampling can provide increased precision compared to simple random sampling, especially if the variation within clusters is smaller than the variation between clusters. Best of all, cluster sampling can reveal detailed information about the specfic cluster compared to the rest of the dataset.

image.png

Follow the steps to perform cluster sampling:
1. Create a list or map of all the clusters in the population.
2. Decide how many clusters you want to include in your sample
3. Use a random sampling technique (e.g., random number generator) to select the required number of clusters from the sampling frame
4. Once clusters are selected, include all elements (individuals, households, etc.) within each chosen cluster in your sample
5. Collect data from the elements within the selected clusters

Let's revisit the credit dataset.

````python
display(df.describe())
````

We can start by sorting the dataset by the income. By sorting the dataset first and then assigning the cluster we are able to group based on nearby incomes.

````python
# Sorting by column 'income'
df.sort_values(by=['income'])
````

We can create a function to perform cluster sampling outlining the steps above.
First, we obtain the number of elements in each cluster group and create a list that assigns the group id to the element. Then, we append the group id to the original dataset and randomly choose a cluster to perform analysis on. We choose to keep the random_state as None because calling the function multiple times will reuse the same instance and will produce different results.

Parameters:
* clusters: number of clusters to make from the dataset
* state: determines random number generation for shuffling the data. (default is None).

````python
# function for extracting systematic samples with useful parameters

def sample_cluster(dataframe, clusters, state = None):
    # define variables
    length = len(dataframe)
    element_max = length / clusters
    print(f'  - elements by cluster: {element_max}')

    cluster_list = []
    cluster_id = 0
    element_count = 0

    # define clusters
    for _ in df.iterrows():
        cluster_list.append(cluster_id)
        element_count += 1
        if element_count > (element_max - 1):
            element_count = 0
            cluster_id += 1
    # add cluster number to the table
    dataframe['cluster'] = cluster_list
    print(' - cluster list')
    print(dataframe['cluster'].value_counts())
    print('')
    rd.seed(state)
    # choose a cluster randomly
    cluster_selected = rd.randint(0, clusters - 1)
    print('cluster selected:',cluster_selected)
    dataframe_clustered_selected = dataframe[dataframe['cluster'] == cluster_selected]
    print('cluster size:',dataframe_clustered_selected.shape[0],'\n')
    #dataframe = dataframe.drop(['cluster'], axis=1)
    return dataframe_clustered_selected
````

With our function defined, we can perform clustered sampling with 200 clusters.

````python
# using the function
df_sample_cluster = sample_cluster(df, 200)

display(df_sample_cluster.describe())
````

````python
fig = plt.figure()
plt.figure(figsize=(100,50))
df2 = pd.concat({'df': df.melt(), 'df_sample_cluster': df_sample_cluster.melt()}, names=['source', 'old_index'])
df2 = df2.reset_index(level=0).reset_index(drop=True)

t = sns.catplot(data=df2, kind='box', col='variable', y='value', x='source', palette='turbo',hue='source', legend = 'brief', height=4, aspect=0.8, col_wrap=6, sharey=False)
t.set(xlabel='', xticks=[]) # remove x labels and ticks
t.add_legend()
````

In this example, we see that the cluster sample's income range is a lot smaller than the complete dataset's income range. This would make sense as the cluster groups are not assigned randomly. This also causes the range of the loan and age to be smaller as well, because the amount of loan given should depend on the income and the income made can be based on the age. Focusing on the default column, we see that the clustered dataset's average is lower than the original dataset but the standard deviations are similar.

Cluster sampling can be a cost-effective and practical approach in situations where individual-level sampling is challenging or resource-intensive. However, it's essential to carefully plan and execute the sampling process to ensure that the results are valid and representative of the entire population.

# Bootstrapping

Bootstrapping is a statistical resampling technique employed to estimate the sampling distribution of a statistic. This method involves repeatedly drawing samples from the observed data with replacement.
Bootstrapping can be useful when you need to gauge the uncertainty or variability tied to a particular statistic, such as mean, median, or regression coefficient, especially when dealing with limited dataset sizes.

Furthermore, bootstrapping finds utility in tasks like model validation and outlier detection. It proves especially advantageous in situations where conducting cross-validation poses challenges or becomes computationally intensive.

image.png

To perform boostrapping effectively, follow these steps:
1. Decide on the number of times to sample (n).
2. Randomly draw a sample that is the same size as the original sample with replacement, meaning the same data point can be selected more than once in the bootstrap sample.
3. Repeat Step 2 to create n bootstrap samples

Let's revisit the credit dataset one more time and remove the cluster column.

````python
df = df.drop(['cluster'], axis=1)
display(df.describe())
````

````python
# Libraries for bootstrap
from sklearn.utils import resample
````

We will use the ```resample()``` function from the utils library to select the sample size to be the length of the credit table. In this example, since we are creating multiple samples, we want to control the results produced. We set the random_state to allow elements to change everytime the function is executed and pass 200 as a seed to generate a random number.

Parameters:
* n: the number samples to be generated
* replace: whether the elements of the table can be inserted into the sample more than once
* random_state: determines random number generation for shuffling the data. (default is None).


````python
# first sample
df_boot1 = resample(df, replace=True, n_samples=len(df), random_state=200)
````

````python
# second sample
df_boot2 =  resample(df, replace=True, n_samples=len(df), random_state=200)
````

````python
# now let's compare
display(df_boot1.describe())
display(df_boot2.describe())
````

````python
# box plot for original vs boot1
fig = plt.figure()
plt.figure(figsize=(100,50))
df2 = pd.concat({'df': df.melt(), 'df_boot1': df_boot1.melt(), 'df_boot2': df_boot2.melt()}, names=['source', 'old_index'])
df2 = df2.reset_index(level=0).reset_index(drop=True)

t = sns.catplot(data=df2, kind='box', col='variable', y='value', x='source', palette='turbo',hue='source', legend = 'brief', height=4, aspect=0.8, col_wrap=6, sharey=False)
t.set(xlabel='', xticks=[]) # remove x labels and ticks
t.add_legend()
````

Comparing the statistics for the original dataset and the bootstrapped dataset, we get very similar results. With the least amount of difference in the mean and standard deviation among the attributes, there may be some error due to the bootstrapped samples being higher than the given dataset.

While bootstrapping is a powerful technique, the quality of the results depends on the quality of the original dataset and the assumptions made about the data. Additionally, the number of bootstrap resamples should be chosen carefully to balance computational resources and accuracy. More resamples generally lead to more accurate estimates but also require more computation time.

#Common data sampling errors
A sampling error is a difference between the sampled value and the true population value. Sampling errors happen during data collection when the sample is not typical of the population or is biased in some way.

Because a sample is merely an approximation of the population from which it is collected, even randomized samples will have some sampling error.

The following are some common data sampling errors:

* **Population-specific error:** A population-specific error occurs when a researcher doesn't understand who to survey.
* **Selection error:** Selection bias arises when the sample is chosen in a way that favors a specific group or trait. For example, if a health study is only conducted on people who are willing to participate, the sample may not be representative of the overall community.
* **Non-response error:** This bias happens when people chosen for the sample do not participate in the survey or study. As a result, certain groups may be underrepresented, affecting the accuracy of the results.
* **Random Sampling Error:** This is the most common type of sampling error and occurs due to the natural variability or chance fluctuations in the data. Even with a random, unbiased sample, you would expect some degree of variation in the sample statistics from one sample to another.
* **Systematic Sampling Error:** Systematic errors occur when there is a consistent bias or problem in the sampling process. For example, if you consistently select every 10th person in a list to be in your sample, and there is some pattern in the list (e.g., the list is sorted by age), then your sample may not be representative of the population.

* **Selection Bias:** Selection bias occurs when certain groups or elements in the population are more or less likely to be included in the sample. This can happen if the sampling method is not truly random or if certain subpopulations are underrepresented.

* **Undercoverage Bias:** Undercoverage bias arises when some parts of the population are not adequately represented in the sample. For example, if you conduct a telephone survey but only reach people with landline phones, you may underrepresent younger individuals who primarily use cell phones.

* **Measurement Error:** Measurement error can occur when the tools or instruments used to collect data introduce inaccuracies. This can include errors in recording responses, interpreting data, or measurement instrument issues.

* **Non-Sampling Error:** While not strictly a type of sampling error, non-sampling error encompasses all other errors that can occur in the research process, including data entry errors, interviewer bias, and response bias. These errors can affect the quality of the data and the accuracy of the estimates.
